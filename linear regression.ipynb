```python
# Step 0: Import libraries
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error

# Step 1: Dataset
# Feature 1: Hours Studied, Feature 2: Hours Slept
X = np.array([[1, 7],
              [2, 6],
              [3, 7],
              [4, 8],
              [5, 5],
              [6, 6],
              [7, 7],
              [8, 8]])
Y = np.array([2, 4, 5, 4, 5, 6, 7, 8])

# Step 2: Train-Test Split (75% train, 25% test)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=42)

# Dictionary to store R² scores
r2_scores = {}

# --------------------------
# Model 1: Simple Linear Regression (Hours Studied only)
# --------------------------
X_train_simple = X_train[:, 0].reshape(-1,1)
X_test_simple = X_test[:, 0].reshape(-1,1)

lin_model = LinearRegression()
lin_model.fit(X_train_simple, Y_train)

Y_test_pred_lin = lin_model.predict(X_test_simple)
r2_scores['Simple Linear'] = r2_score(Y_test, Y_test_pred_lin)

print("=== Simple Linear Regression ===")
print(f"Equation: y = {lin_model.intercept_:.2f} + {lin_model.coef_[0]:.2f}*x")
print("Testing R²:", r2_scores['Simple Linear'])
print("Testing MSE:", mean_squared_error(Y_test, Y_test_pred_lin))
print()

# --------------------------
# Model 2: Multiple Linear Regression (Hours Studied + Hours Slept)
# --------------------------
multi_model = LinearRegression()
multi_model.fit(X_train, Y_train)

Y_test_pred_multi = multi_model.predict(X_test)
r2_scores['Multiple Linear'] = r2_score(Y_test, Y_test_pred_multi)

print("=== Multiple Linear Regression ===")
print(f"Equation: y = {multi_model.intercept_:.2f} + {multi_model.coef_}*X")
print("Testing R²:", r2_scores['Multiple Linear'])
print("Testing MSE:", mean_squared_error(Y_test, Y_test_pred_multi))
print()

# --------------------------
# Model 3: Polynomial Regression (degree 2, Hours Studied only)
# --------------------------
poly = PolynomialFeatures(degree=2)
X_train_poly = poly.fit_transform(X_train_simple)
X_test_poly = poly.transform(X_test_simple)

poly_model = LinearRegression()
poly_model.fit(X_train_poly, Y_train)

Y_test_pred_poly = poly_model.predict(X_test_poly)
r2_scores['Polynomial (deg 2)'] = r2_score(Y_test, Y_test_pred_poly)

print("=== Polynomial Regression (degree 2) ===")
print("Testing R²:", r2_scores['Polynomial (deg 2)'])
print("Testing MSE:", mean_squared_error(Y_test, Y_test_pred_poly))
print()

# --------------------------
# Select best model
# --------------------------
best_model_name = max(r2_scores, key=r2_scores.get)
print(f"Best model according to testing R² score: {best_model_name} with R² = {r2_scores[best_model_name]:.2f}")
```
